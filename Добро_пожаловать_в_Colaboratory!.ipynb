```
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

h = 0.01
def euler (dydx, x, y0, n):
    y = [y0]
    for i in range(n-1):
      xi = x[i]
      yi = y[i]
      yi1 = yi + h * dydx(yi, xi)
      #x.append(xi + h)
      y.append(yi1)
    return y

def dydx(y, x):
    k = 0.42
    return k * y

x0, y0 = 0,1

n = 10000
x = np.arange(0,10+h,h)
t = np.linspace(0,10,1000)

y = euler(dydx, x=x,y0=y0, n = len(x - 1))
solve = odeint(dydx, y0=y0, t=t)
solve = solve.flatten()

plt.plot(x,y,label=r"euler")
plt.plot(t,solve,label=r"x'=k(x)")
plt.grid(True)
plt.legend()
plt.show()

```
