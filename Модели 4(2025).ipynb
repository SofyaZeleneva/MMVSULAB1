логистическое отображение

```python
import matplotlib.pyplot as plt
import math

# Параметры
n_skip = 10  # Количество итераций
x0 = 0.5      # Начальное значение
r = 1 + math.sqrt(6)      # изменяется (0,4)

# Инициализация списка (начинаем с x0)
x = [x0]

for n in range(n_skip):
    next_x = r * x[-1] * (1 - x[-1])  # Берём последнее значение x
    x.append(next_x)

plt.figure(figsize=(8, 5))
plt.plot(x, 'b-', linewidth=1, label=f"r = {r}")

plt.title(f"Итерации отображения")
plt.ylabel("$X_n$")
plt.grid(True, alpha=0.5)
plt.legend()  # Теперь легенда будет отображаться
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sine_map(r, x0, n_iter=10, n_skip=10):
    """
    Параметры:
    - r: параметр управления
    - x0: начальное значение
    - n_iter: количество итераций
    - n_skip: количество точек, которые нужно пропустить (чтобы уйти от переходного процесса)

    Возвращает:
    - Массив значений Xₙ
    """
    x = x0
    trajectory = []

    # Пропускаем первые n_skip итераций для устранения переходного процесса
    for _ in range(n_skip):
        x = r * x * (1 - x)

    # Сохраняем следующие n_iter точек
    for _ in range(n_iter):
        x = r * x * (1 - x)
        trajectory.append(x)

    return np.array(trajectory)

def plot_sine_map(r_values, x0=0.2, n_iter=1000, n_skip=100):
    """
    Строит бифуркационную диаграмму для отображения синуса.

    Параметры:
    - r_values: массив значений параметра r
    - x0: начальное значение
    - n_iter: количество итераций для каждого r
    - n_skip: количество точек, которые нужно пропустить
    """
    plt.figure(figsize=(10, 6))

    for r in r_values:
        trajectory = sine_map(r, x0, n_iter, n_skip)
        plt.plot([r] * len(trajectory), trajectory, ',k', alpha=0.5, markersize=1)

    plt.title("Бифуркационная диаграмма")
    plt.xlabel("Параметр $r$")
    plt.ylabel("$X_n$")
    plt.grid(True, alpha=0.3)
    plt.show()

if __name__ == "__main__":
    # r (от 0 до 4 с шагом 0.001)
    r_values = np.linspace(0, 4, 1000)
    plot_sine_map(r_values)
```

Лестница Ламерея

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(x, r):
    return r * x * (1 - x)

def plot_lameere_ladder(r, x0, n_iter=50):
    x = np.linspace(0, 1, 1000)
    y = logistic_map(x, r)

    # Траектория
    trajectory = [x0]
    for _ in range(n_iter):
        trajectory.append(logistic_map(trajectory[-1], r))

    # Построение
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, 'b-', label=f'$f(x) = {r}x(1-x)$')
    plt.plot(x, x, 'k--', label='$y = x$')

    # Лестница Ламерея
    for i in range(n_iter):
        xn = trajectory[i]
        xn1 = trajectory[i+1]
        plt.plot([xn, xn], [xn, xn1], 'r-', lw=1)  # Вертикальная линия
        plt.plot([xn, xn1], [xn1, xn1], 'r-', lw=1)  # Горизонтальная линия

    plt.title(f'Лестница Ламерея для $r = {r}$')
    plt.xlabel('$x_n$')
    plt.ylabel('$x_{n+1}$')
    plt.legend()
    plt.grid()
    plt.show()

# Примеры для разных r:
plot_lameere_ladder(r=1, x0=0.1)
plot_lameere_ladder(r=2, x0=0.1)
plot_lameere_ladder(r=3, x0=0.1)
plot_lameere_ladder(r=3.5, x0=0.1)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sine_map(r, x0, n_iter=1000, n_skip=100):
    """
    Параметры:
    - r: параметр управления
    - x0: начальное значение
    - n_iter: количество итераций
    - n_skip: количество точек, которые нужно пропустить (чтобы уйти от переходного процесса)

    Возвращает:
    - Массив значений Xₙ
    """
    x = x0
    trajectory = []

    # Пропускаем первые n_skip итераций для устранения переходного процесса
    for _ in range(n_skip):
        x = r * np.sin(np.pi * x)

    # Сохраняем следующие n_iter точек
    for _ in range(n_iter):
        x = r * np.sin(np.pi * x)
        trajectory.append(x)

    return np.array(trajectory)

def plot_sine_map(r_values, x0=0.2, n_iter=500, n_skip=200):
    """
    Строит бифуркационную диаграмму для отображения синуса.

    Параметры:
    - r_values: массив значений параметра r
    - x0: начальное значение
    - n_iter: количество итераций для каждого r
    - n_skip: количество точек, которые нужно пропустить
    """
    plt.figure(figsize=(10, 6))

    for r in r_values:
        trajectory = sine_map(r, x0, n_iter, n_skip)
        plt.plot([r] * len(trajectory), trajectory, ',k', alpha=0.5, markersize=1)

    plt.title("Бифуркационная диаграмма")
    plt.xlabel("Параметр $r$")
    plt.ylabel("$X_n$")
    plt.grid(True, alpha=0.3)
    plt.show()

if __name__ == "__main__":
    # r (от 0.2 до 1 с шагом 0.001)
    r_values = np.linspace(0.2, 1, 1000)
    plot_sine_map(r_values)
```

Отображение Гауса

```python
import matplotlib.pyplot as plt
import math

# Параметры системы
a = 4.9       # Параметр в экспоненте
r = 0.5       # изменяется (-1,1)
x0 = 0.1      # Начальное значение
n_skip = 1000 # Число итераций

# Инициализация списка значений
x = [x0]

# Итеративное вычисление
for n in range(n_skip):
    next_x = math.exp(-a * x[-1]**2) + r
    x.append(next_x)

# Построение графика
plt.figure(figsize=(10, 5))
plt.plot(x, 'b-', linewidth=1, label=f"r = {r}, a = {a}")

# Оформление
plt.title(f"Итерации отображения $X_{{n+1}} = e^{{-{a}X_n^2}} + {r}$")
plt.xlabel("Номер итерации $n$")
plt.ylabel("$X_n$")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

Отображение Гаусса Бифуркационная диаграмма

```python
import numpy as np
import matplotlib.pyplot as plt
import math

a = 4.9
n_skip = 500  # Пропускаем переходный процесс
n_iter = 10  # Число точек для каждого r
x0 = 0.1

r_values = np.linspace(-1, 1, 1000)  # Диапазон r

plt.figure(figsize=(12, 6))

for r in r_values:
    x = x0
    # Пропускаем переходный процесс
    for _ in range(n_skip):
        x = math.exp(-a * x**2) + r
    # Записываем точки для построения
    for _ in range(n_iter):
        x = math.exp(-a * x**2) + r
        plt.plot(r, x, ',k', alpha=0.5, markersize=1)

plt.title(f"Бифуркационная диаграмма: $X_{{n+1}} = e^{{-{a}X_n^2}} + r$")
plt.xlabel("Параметр $r$")
plt.ylabel("$X_n$ (устойчивые состояния)")
plt.grid(True, alpha=0.3)
plt.show()
```

показатели Ляпуновского

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

# Настройка графики
plt.rcParams['figure.figsize'] = [12, 10]
plt.rcParams['font.size'] = 12

def lyapunov_exponent(f, df, r, x0=0.5, n_transient=1000, n_iter=5000):
    """Вычисление показателя Ляпунова для одного значения параметра r"""
    x = x0
    # Пропускаем переходный процесс
    for _ in range(n_transient):
        x = f(x, r)

    # Вычисляем показатель Ляпунова
    lyap = 0.0
    for _ in range(n_iter):
        x = f(x, r)
        lyap += np.log(abs(df(x, r)))

    return lyap / n_iter

# Определение отображений и их производных
def logistic_map(x, r):
    return r * x * (1 - x)

def d_logistic_map(x, r):
    return r * (1 - 2*x)

def sine_map(x, r):
    return r * np.sin(np.pi * x)

def d_sine_map(x, r):
    return r * np.pi * np.cos(np.pi * x)

def gauss_map(x, r, a=4.9):
    return np.exp(-a * x**2) + r

def d_gauss_map(x, r, a=4.9):
    return -2 * a * x * np.exp(-a * x**2)

def planck_map(x, r):
    return (r * x**3) / (1 + np.exp(x))

def d_planck_map(x, r):
    return (r * x**2 * (3*(1 + np.exp(x)) - x*np.exp(x))) / (1 + np.exp(x))**2

# Параметры для расчетов
r_logistic = np.linspace(2.8, 4.0, 1000)
r_sine = np.linspace(0.2, 1.0, 1000)
r_gauss = np.linspace(-1.0, 1.0, 10)
r_planck = np.linspace(1.0, 7.0, 10)

# Вычисление показателей Ляпунова
lyap_logistic = [lyapunov_exponent(logistic_map, d_logistic_map, r) for r in tqdm(r_logistic, desc="Logistic Map")]
lyap_sine = [lyapunov_exponent(sine_map, d_sine_map, r) for r in tqdm(r_sine, desc="Sine Map")]
lyap_gauss = [lyapunov_exponent(gauss_map, d_gauss_map, r) for r in tqdm(r_gauss, desc="Gauss Map")]
lyap_planck = [lyapunov_exponent(planck_map, d_planck_map, r) for r in tqdm(r_planck, desc="Planck Map")]

# Построение графиков
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# Логистическое отображение
axes[0,0].plot(r_logistic, lyap_logistic, 'b-', lw=0.5)
axes[0,0].axhline(0, color='red', linestyle='--', alpha=0.5)
axes[0,0].set_title('Логистическое отображение')
axes[0,0].set_xlabel('Параметр r')
axes[0,0].set_ylabel('Показатель Ляпунова')
axes[0,0].grid(True, alpha=0.3)

# Синус-отображение
axes[0,1].plot(r_sine, lyap_sine, 'g-', lw=0.5)
axes[0,1].axhline(0, color='red', linestyle='--', alpha=0.5)
axes[0,1].set_title('Синус-отображение')
axes[0,1].set_xlabel('Параметр r')
axes[0,1].grid(True, alpha=0.3)

# Отображение Гаусса
axes[1,0].plot(r_gauss, lyap_gauss, 'r-', lw=0.5)
axes[1,0].axhline(0, color='red', linestyle='--', alpha=0.5)
axes[1,0].set_title('Отображение Гаусса')
axes[1,0].set_xlabel('Параметр r')
axes[1,0].set_ylabel('Показатель Ляпунова')
axes[1,0].grid(True, alpha=0.3)

# Отображение Планка
axes[1,1].plot(r_planck, lyap_planck, 'm-', lw=0.5)
axes[1,1].axhline(0, color='red', linestyle='--', alpha=0.5)
axes[1,1].set_title('Отображение Планка')
axes[1,1].set_xlabel('Параметр r')
axes[1,1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
