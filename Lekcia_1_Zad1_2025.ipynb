```python
import numpy as np
import matplotlib.pyplot as plt

def fibonacci_model(N, X0, X1):
    seq = np.zeros(N)
    seq[0], seq[1] = X0, X1
    for n in range(2, N):
        seq[n] = seq[n-1] + seq[n-2]
    return seq

# Параметры
N = 20
X0, X1 = 1, 1

# Расчет и график
fib_seq = fibonacci_model(N, X0, X1)

plt.figure(figsize=(10, 5))
plt.plot(fib_seq, 'bo-', label=f'X0={X0}, X1={X1}')
plt.title('Модель Фибоначчи: $X_{n+2} = X_{n+1} + X_n$')
plt.xlabel('n')
plt.ylabel('X_n')
plt.grid(True)
plt.legend()
plt.show()
```

Случай а) R = 1/2 * r

```python
def model_a(N, U0, U1, r):
    R = 0.5 * r
    seq = np.zeros(N)
    seq[0], seq[1] = U0, U1
    for n in range(2, N):
        seq[n] = (2 + R/r)*seq[n-1] - seq[n-2]
    return seq

# Параметры
N = 30
U0, U1 = 1, 1
r = 0.5

# Расчет и график
seq_a = model_a(N, U0, U1, r)

plt.figure(figsize=(10, 5))
plt.plot(seq_a, 'ro-', label=f'R = r/2 (r={r})')
plt.title('Модель: $U_{n+2} - (2+R/r)U_{n+1} + U_n = 0$')
plt.xlabel('n')
plt.ylabel('U_n')
plt.grid(True)
plt.legend()
plt.show()
```

Случай б) r = 1/2 * R

```python
def model_b(N, U0, U1, R):
    r = 0.5 * R
    seq = np.zeros(N)
    seq[0], seq[1] = U0, U1
    for n in range(2, N):
        seq[n] = (2 + R/r)*seq[n-1] - seq[n-2]
    return seq

# Параметры
N = 30
U0, U1 = 1, 1
R = 0.5

# Расчет и график
seq_b = model_b(N, U0, U1, R)

plt.figure(figsize=(10, 5))
plt.plot(seq_b, 'gs-', label=f'r = R/2 (R={R})')
plt.title('Модель: $U_{n+2} - (2+R/r)U_{n+1} + U_n = 0$')
plt.xlabel('n')
plt.ylabel('U_n')
plt.grid(True)
plt.legend()
plt.show()
```

Для модели Фибоначчи используется простая итеративная формула

Во второй модели:

В случае а) сначала задается r, затем вычисляется R = r/2

В случае б) сначала задается R, затем r = R/2

Коэффициент (2 + R/r) автоматически пересчитывается в каждом случае

Все графики имеют подписи и легенды для наглядности

Вы можете легко менять начальные условия (X0, X1, U0, U1) и параметры (r, R) для исследования поведения моделей.
