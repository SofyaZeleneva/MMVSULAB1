построить графики для следующих функций:
1) X1(k) = 1/9 * 24 * 2^k - 1/9 * 24 * (-1)^k - 4*k(-1)^k
2) X2(k) = 1/9 * 4 * 2^k - 1/9 * 4 * (-1)^k + 1/3*4*k(-1)^k
3) X3(k) = 1/9 * 2^k - 1/9 * 8 * (-1)^k + 1/3*(-2)*k(-1)^k

```python
import numpy as np
import matplotlib.pyplot as plt

# Создаем массив значений k
k = np.arange(0, 10, 1)  # от 0 до 9 с шагом 1

# Вычисляем значения функций
X1 = (1/9)*24*2**k - (1/9)*24*(-1)**k - 4*k*(-1)**k
X2 = (1/9)*4*2**k - (1/9)*4*(-1)**k + (1/3)*4*k*(-1)**k
X3 = (1/9)*2**k - (1/9)*8*(-1)**k + (1/3)*(-2)*k*(-1)**k

# Настраиваем размер графика
plt.figure(figsize=(12, 8))

# Строим графики
plt.plot(k, X1, 'bo-', linewidth=2, markersize=8, label='$X_1(k) = \\frac{24}{9}2^k - \\frac{24}{9}(-1)^k - 4k(-1)^k$')
plt.plot(k, X2, 'rs--', linewidth=2, markersize=8, label='$X_2(k) = \\frac{4}{9}2^k - \\frac{4}{9}(-1)^k + \\frac{4}{3}k(-1)^k$')
plt.plot(k, X3, 'g^-.', linewidth=2, markersize=8, label='$X_3(k) = \\frac{2}{9}^k - \\frac{8}{9}(-1)^k - \\frac{2}{3}k(-1)^k$')

# Добавляем подписи и легенду
plt.title('Графики функций X1(k), X2(k) и X3(k)', fontsize=14)
plt.xlabel('k', fontsize=12)
plt.ylabel('Значение функции', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')

# Отображаем значения в точках
for i in range(len(k)):
    plt.text(k[i], X1[i], f'{X1[i]:.1f}', ha='center', va='bottom')
    plt.text(k[i], X2[i], f'{X2[i]:.1f}', ha='center', va='bottom')
    plt.text(k[i], X3[i], f'{X3[i]:.1f}', ha='center', va='top')

plt.tight_layout()
plt.show()
```
